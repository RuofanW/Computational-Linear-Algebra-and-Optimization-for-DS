```python
import numpy as np
import pandas as pd
```

```python
A = np.array([[2, -1], [-1, 2]])
```

```python
u0 = np.array([[1],[0]])
```

```python
u = {}
u[0] = u0
for i in range(3):
    u0 = np.dot(A, u0)
    u[i + 1] = u0
```

As k increases, $u_k$ will converge to a multiple of $(-1, 1)$, which is one of the eigenvector of A

### More exploration and verification

```python
u = {}
u[0] = u0
for i in range(100):
    u0 = np.dot(A, u0)
    u[i + 1] = u0
```

```python
data = {'index':[], 'u1':[], 'u2':[]}
for i in range(100):
    data['index'].append(i)
    data['u1'].append(float(u[i][0]))
    data['u2'].append(float(u[i][1]))
```

```python
data = pd.DataFrame(data)
```
