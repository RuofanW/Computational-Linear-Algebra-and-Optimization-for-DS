```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
uc = np.array([[np.cos(theta),np.sin(theta)] for theta in np.linspace(0,2*np.pi,128)]).reshape((128,2))
```

```python
uc
```

```python
A = np.array([[1,3],
              [2,3]],dtype = "float64")
u,sv,vt = np.linalg.svd(A)
fig,ax = plt.subplots(1,2,figsize = (8,3.5))
ax[0].plot(uc[:,0],uc[:,1],"r.")
ax[0].set_title("unit circle")
ax[1].plot((uc@A)[:,0],(uc@A)[:,1],"b.")
ax[1].set_title("unit circle under A")
```

## dimension reduction via SVD

```python
import scipy.ndimage as ni
import matplotlib.image as mpimg
```

```python
img=ni.imread("img.png",flatten = 1)
```

```python
plt.imshow(img,cmap = 'gray')
```

```python
u,s,v = np.linalg.svd(img)
```

```python
plt.clf()
l = len(list(range(0,375,30)))
fig2,ax2 = plt.subplots(l,figsize = (40,60))
for ind,i in enumerate(range(0,375,30)):
    ax2[ind].set_title("low rank = " +str(i))
    ax2[ind].imshow(u[:,:i]@np.diag(s[:i])@(v[:i,:]),cmap = 'gray')
```
